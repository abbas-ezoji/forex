```python
for c in all_coins:
    symbol = c['symbol']
    data_url = f'https://api.lunarcrush.com/v2/assets?data=assets&data_points=720&symbol={symbol}&interval=day&key={api_key}'
    print(data_url)
    r = requests.get(data_url)
    data=r.json()
    timeSeries = data['data'][0]['timeSeries']
    df_timeSeries = pd.DataFrame.from_records(timeSeries)
    df_timeSeries.to_csv(f'timeSeries_{symbol}.csv')
```
