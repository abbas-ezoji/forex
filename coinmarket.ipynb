```python
import requests
from bs4 import BeautifulSoup
```

```python
base_url = 'https://coinmarketcal.com/en/event/staking-3-0-launch-52823'
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
r = requests.get(base_url, headers=headers)
soup = BeautifulSoup(r.text, 'html.parser')
soup
```

```python
ev_list_sec = soup.find('div', {'id' : 'chartdiv'})
all_data = ev_list_sec['data-chart']
event_date = ev_list_sec['data-event']
```

```python
import json
```

```python
all_data=json.loads(all_data)
event_date=json.loads(event_date)
```

```python
import pandas as pd

df_all_data = pd.DataFrame.from_records(all_data)
df_event_data = pd.DataFrame.from_records(event_date)
```

```python
df_all_data.to_csv('all_data.csv')
df_event_data.to_csv('event_data.csv')
```
